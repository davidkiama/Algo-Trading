```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('GOOG.csv', index_col='Date', parse_dates=True)
goog_data = df.loc['2014-01-01':'2018-01-01']
```

```python
goog_data = goog_data.tail(620)
lows = goog_data['Low']
highs = goog_data['High']
```

```python
fig = plt.figure(figsize=(14,10))
ax1=fig.add_subplot(111, ylabel='Google Price in $')
highs.plot(ax=ax1, color='c', lw=2)
lows.plot(ax=ax1, color='y', lw=2)
plt.hlines(highs.head(200).max(), lows.index.values[0], lows.index.values[-1], color='g', lw=3)
plt.hlines(lows.head(200).min(), lows.index.values[0], lows.index.values[-1],color='r', lw=2)
plt.axvline(color='b', lw=2, x=lows.index.values[200], ls=':')
```
