```python
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
goog_data = pd.read_csv('GOOG.csv', index_col='Date', parse_dates=True)
```

```python
goog_monthly_return = goog_data['Adj Close'].pct_change().groupby(\
    [goog_data['Adj Close'].index.year, \
    goog_data['Adj Close'].index.month]).mean()
goog_monthly_return_list =[]
```

```python
for i in range(len(goog_monthly_return)):
    goog_monthly_return_list.append\
        ({'month':goog_monthly_return.index[i][1],
        'monthly_return':goog_monthly_return.iloc[i]})
```

```python
goog_monthly_return_list=pd.DataFrame(goog_monthly_return_list,columns=('month','monthly_return'))
```

```python
goog_monthly_return_list.boxplot(column='monthly_return', by='month', figsize=(14,10))
ax = plt.gca()
labels = [item.get_text() for item in ax.get_xticklabels()]
labels=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
ax.set_xticklabels(labels)
ax.set_ylabel('GOOG return')
plt.tick_params(axis='both', which='major', labelsize=7)
plt.title("GOOG Monthly return 2010-2019")
plt.suptitle("")
```
